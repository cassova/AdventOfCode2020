You glance back down at your bag and try to remember why you brought so many adapters; there must be more than a trillion valid ways to arrange them! Surely, there must be an efficient way to count the arrangements.

What is the total number of distinct ways you can arrange the adapters to connect the charging outlet to your device?

Your puzzle answer was `24803586664192`.

```python
import numpy as np

def day10():
    data = [int(x) for x in open('./assets/advent2020_d10.txt').readlines()]
    data.extend([0,max(data)+3])
    data = np.array(sorted(data))

    x = 0
    d = {1:0, 2:0, 3:0}
    for i,v in enumerate(data):
        if i>0:
            n=v-x
            d[n]+=1
        x=v
    part1 = d[1]*d[3]


    # Part 2 - Got help from Sharon Sung and Naga to come up with this solution
    
    total = 1
    cur_len = 0
    factors = {1:1, 2:2, 3:4, 4:7}
    for x in np.diff(data):
        if x==1:
            cur_len += 1
        elif cur_len > 0:
            total *= factors[cur_len]
            cur_len=0
    part2 = total
    
    return part1, part2
```

```python
# Display Results
part1,part2 = day10()
print(f'Part 1 Answer = {part1}')
print(f'Part 2 Answer = {part2}')
```
